```python
import pandas as pd

csv_df = pd.read_csv(r"E:\Code_Projects\Brokers\fyers\candles_1m.csv")

print(csv_df.columns.tolist())
print(csv_df.head())
```

```python
mismatch_summary = {}

for symbol, api_df in data.items():
    csv_symbol_df = csv_df[csv_df["symbol"] == symbol]

    if csv_symbol_df.empty:
        mismatch_summary[symbol] = "missing_in_csv"
        continue

    csv_symbol_df = csv_symbol_df.sort_values("bucket_epoch").reset_index(drop=True)
    api_df = api_df.sort_values("timestamp").reset_index(drop=True)

    merged = csv_symbol_df.merge(
        api_df,
        left_on="bucket_epoch",
        right_on="timestamp",
        suffixes=("_csv", "_api"),
    )

    mismatch = merged[
        (merged["open_csv"] != merged["open_api"]) |
        (merged["high_csv"] != merged["high_api"]) |
        (merged["low_csv"] != merged["low_api"]) |
        (merged["close_csv"] != merged["close_api"])
    ]

    mismatch_summary[symbol] = len(mismatch)

print(mismatch_summary)
```
